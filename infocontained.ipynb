```python
import pandas as pd
import spacy
spacy.prefer_gpu()
nlp = spacy.load("en_core_web_sm")


datafilepath = "./data/"
```

```python
def analysis(listOfTopicNumbers):
    for eachTopic in listOfTopicNumbers:
        topicFile = datafilepath + str(eachTopic)+'.tsv'
        topicFile_df = pd.read_csv(topicFile, sep="\t", header=None)

        # added a column to the nlp linguistic annotation features
        topicFile_df["doc"] = topicFile_df[1].apply(nlp)

    display(topicFile_df)
        

```

```python
if __name__ == "__main__":
    

    analysis([1])

    # analysis([x for x in range(1,81)])
```
